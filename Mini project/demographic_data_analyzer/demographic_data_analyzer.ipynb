# 8. What country has the highest percentage of people that earn >50K and what is that percentage?

```python
df.head()
```

```python
country_morethan50k = df[df['salary'] == '>50K']['native-country'].value_counts()
country_morethan50k
```

```python
# Mengeluarkan nilai tertinggi pada list
countrybest_morethan50k = country_morethan50k.idxmax()
countrybest_morethan50k
```

```python
percentage_countrybest_morethan50k = int(country_morethan50k.max())/len(df)*100
print('Country has the highest percentage of people that earn >50K and what is that percentage: ', percentage_countrybest_morethan50k, '%')
```
